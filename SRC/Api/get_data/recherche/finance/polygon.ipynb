```python
parent_dir = os.path.abspath(os.path.join("../../../", 'API_KEYS.json'))
with open(parent_dir, "r") as fichier:
    # Charger le contenu du fichier JSON dans une variable
    data = json.load(fichier)
```

```python
API_KEY = data["API"][2]["PolygonIo"]
```

```python
from polygon import RESTClient

# Clé API Polygon.io
client = RESTClient(API_KEY)

# Recherche des options de Tesla
contracts = client.list_options_contracts(underlying_ticker="TSLA")  # Utilisation correcte du paramètre

# Résumé des options
for contract in contracts:
    print(f"Option: {contract.ticker}")
    print(f"  - Type : {contract.contract_type}")
    print(f"  - Strike : {contract.strike_price}")
    print(f"  - Expiration : {contract.expiration_date}")
    # print(f"  - Last Price : {contract.last_price}\n")
```

```python
from polygon import RESTClient

# Remplacez par votre clé API Polygon.io
client = RESTClient(API_KEY)

# Récupération des 10 premiers contrats disponibles
contracts = client.list_options_contracts(limit=10)

# Filtrage des contrats liés à Tesla
for contract in contracts:
    if 'TSLA' in contract.ticker:  # Vérifie si TSLA est dans le ticker
        print(f"Option: {contract.ticker}")
        print(f"  - Type : {contract.contract_type}")
        print(f"  - Strike : {contract.strike_price}")
        print(f"  - Expiration : {contract.expiration_date}")
        print(f"  - Last Price : {contract.last_price}\n")
```
