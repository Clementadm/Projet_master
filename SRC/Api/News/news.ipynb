**language=fr** pour obtenir uniquement des articles en français.  

```python
def get_news(business_name, anteriorite_en_jours) :
    from_date = (datetime.now() - timedelta(days=anteriorite_en_jours)).strftime('%Y-%m-%d')  # Exemple : articles des 7 derniers jours
    url = f"https://newsapi.org/v2/everything?q={business_name}&from={from_date}&sortBy=popularity&apiKey={API_KEY}"
    response = requests.get(url)
    data = response.json()
    if data["status"] == "ok":
        return data
    else :
        return 'Status Not ok'
```

```python
all_news = get_news(
    business_name = "Sanofi", 
    anteriorite_en_jours = 7
)
```

```python
import requests


from_date = (datetime.now() - timedelta(days=7)).strftime('%Y-%m-%d')  # Exemple : articles des 7 derniers jours

url = f"https://newsapi.org/v2/everything?q={query}&from={from_date}&sortBy=popularity&apiKey={API_KEY}"

response = requests.get(url)
data = response.json()

if data["status"] == "ok":
    for article in data["articles"]:
        print(f"Title: {article['title']}")
        print(f"Published At: {article['publishedAt']}")
        print(f"Source: {article['source']['name']}")
        print(f"Description: {article['description']}\n")
else:
    print("Erreur:", data.get("message"))
```
