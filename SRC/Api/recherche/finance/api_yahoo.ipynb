```python
import yfinance as yf
from typing import Literal
```

```python
def get_data(business_name : str, period_choose : str)-> None :
    business = yf.Ticker(business_name)
    trade_hist = business.history(period=period_choose)
    analyst_price_targets = business.analyst_price_targets
    grade = business.upgrades_downgrades
    recommendation = business.recommendations
    return trade_hist, analyst_price_targets, grade, recommendation
```

```python
trade_hist, analyst_price_targets, grade, recommendation = get_data("AAPL", "3mo")
```

```python
# 1. Catégorie Positive : Ces termes indiquent une perspective optimiste pour l’action, suggérant d’acheter ou de surpondérer :
# 2. Catégorie Neutre : Ces termes indiquent une perspective équilibrée, conseillant souvent de garder ou de ne pas agir de manière significative :
# 3. Catégorie Négative : Ces termes reflètent une opinion pessimiste, recommandant de vendre ou de réduire l’exposition :

# Les recommandations positives encouragent les investisseurs à acheter ou à renforcer leur position.
##  Les recommandations neutres suggèrent d’attendre sans prendre de décision précipitée.
# Les recommandations négatives avertissent que l’action pourrait sous-performer et indiquent souvent de vendre ou de réduire sa position.

grade["ToGrade"] = grade["ToGrade"].replace([
    "Buy", "Strong Buy", "Outperform", "Overweight", "Neutral", "Hold", "Market Perform", "Equal-Weight", "Sector Perform", "Perform", "Sell", "Underperform", "Underweight"],
    ["Positif", "Positif", "Positif", "Positif", "Neutre", "Neutre", "Neutre", "Neutre", "Neutre", "Neutre", "Negative", "Negative", "Negative"]
)

grade["ToGrade"] = grade["ToGrade"].replace(
    ["Positif", "Neutre", "Negative"],
    [1, 0, -1])

grade = grade.reset_index()
```

```python
# get option chain for specific expiration
opt = msft.option_chain('2024-11-22')
```

L'acheteur d'une option call dispose du droit d'acheter un actif sous-jacent pendant une période donnée à un prix convenu. L'acheteur d'une option put dispose du droit de vendre un actif sous-jacent pendant une période donnée à un prix convenu.

- **lastTradeDate** : Date et heure de la dernière transaction effectuée pour ce contrat d'option.
- **strike** : Le prix d'exercice de l'option. Si vous achetez un call, vous avez le droit d'acheter l'action à ce prix à l'échéance. Si vous achetez un put, vous avez le droit de vendre l'action à ce prix.
- **Bid** : Le prix maximum qu'un acheteur est prêt à payer pour l'option.
- **Ask** : Le prix minimum qu'un vendeur est prêt à accepter pour l'option.
- **Change** : Variation absolue du prix de l'option par rapport à la clôture précédente.
- **PercentChange** : Variation en pourcentage.
volume** : Nombre total de contrats échangés pour cette option au cours de la journée.
- **openInterest** : Nombre total de contrats en circulation (qui n'ont pas été clôturés ou exercés). Cela mesure l'intérêt des investisseurs pour ce contrat.

```python
option_ventes = opt.puts
```
